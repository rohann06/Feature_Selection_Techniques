# 4-Feature Selection-Information gain - mutual information In Regression

```python
import pandas as pd

data = pd.read_csv('Houseprice.csv')
data.head()
```

```python
data.info()
```

```python
numeric_lst=['int64','float64']
numerical_cols = list(data.select_dtypes(include=numeric_lst).columns)
```

```python
numerical_cols
```

```python
data = data[numerical_cols]
data.head()
```

```python
data = data.drop("Id",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.drop(labels=['SalePrice'], axis=1), data['SalePrice'],test_size=0.3,random_state=0)

X_train
```

```python
X_train.isnull().sum()
```

```python
from sklearn.feature_selection import mutual_info_regression
# determine the mutual information
mutual_info = mutual_info_regression(X_train.fillna(0), y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
```

```python
from sklearn.feature_selection import SelectPercentile

### select top 20 
selected_top_columns = SelectPercentile(mutual_info_regression, percentile=20)
selected_top_columns.fit(X_train.fillna(0), y_train)
```

```python
selected_top_columns.get_support()
```
